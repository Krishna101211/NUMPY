Copy:
A copy creates a new array with its own independent data.
It resides in a separate memory location from the original array.
Changes made to the copy will not affect the original array, and vice-versa.
You explicitly create a copy using the .copy() method or by performing operations that inherently create copies (e.g., advanced indexing).

View:
A view creates a new array object that shares the same data as the original array.
It essentially provides a different perspective or "window" into the same memory location as the original array.
Changes made to the view will affect the original array, and vice-versa, because they are referencing the same underlying data.
Views are often created implicitly by certain NumPy operations like slicing or reshaping, and explicitly using the .view() method.
