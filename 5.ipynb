```python
k=np.setxor1d(a,b)
k
```

#**for 2-dimension**

```python
f=np.array([[4,3,2],[5,3,9]])
f
```

```python
s=f.sum()
S=f.sum(axis=0)#add only based on column
Ss=f.sum(axis=1)
s,S,Ss
```

```python
g=f.prod()
G=f.prod(axis=0)
Gg=f.prod(axis=1)
g,G,Gg
```

```python
v=f.mean()
V=f.mean(axis=0)
Vv=f.mean(axis=1)
v,V,Vv
```

```python
n=f.std()
N=f.std(axis=0)
Nn=f.std(axis=1)
n,N,Nn
```

```python
j=f.var()
J=f.var(axis=0)
Jj=f.var(axis=1)
j,J,Jj
```

```python
m=f.min()
M=f.min(axis=0)
Mm=f.min(axis=1)
m,M,Mm
```

```python
l=f.max()
L=f.max(axis=0)
Ll=f.max(axis=1)
l,L,Ll
```

```python
h=f.argmin()
H=f.argmin(axis=0)
Hh=f.argmin(axis=1)
h,H,Hh
```

```python
k=f.argmax()
K=f.argmax(axis=0)
Kk=f.argmax(axis=1)
k,K,Kk
```

```python
p=f.cumsum()
P=f.cumsum(axis=0)
Pp=f.cumsum(axis=1)
p,P,Pp
```

```python
r=f.cumprod()
R=f.cumprod(axis=0)
Rr=f.cumprod(axis=1)
r,R,Rr
```
